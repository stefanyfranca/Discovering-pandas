```python
import pandas as pd
import json

# Supondo que você tenha uma lista de dicionários (o que vai para o seu JSON)
dados_do_estagio = [
    {"pergunta": "Como resetar senha?", "categoria": "Suporte", "tokens": 15},
    {"pergunta": "Qual o plano pro?", "categoria": "Vendas", "tokens": 12},
    {"pergunta": "Erro no login", "categoria": "Suporte", "tokens": 8},
    # ... imagine centenas aqui
]

# 1. Transforma em DataFrame (O coração do Data Science)
df = pd.DataFrame(dados_do_estagio)

# 2. Análise Instantânea: Qual categoria aparece mais?
print("--- Frequência por Categoria ---")
print(df['categoria'].value_counts())

# 3. Estatística Básica: Qual a média de tokens das perguntas?
print(f"\nMédia de tokens: {df['tokens'].mean()}")
```
