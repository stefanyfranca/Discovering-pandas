```python
cols_to_fill_nao_informado = ['Método de pagamento', 'Local', 'Item']
cols_to_numeric_median = ['Preço por unidade', 'Total Gasto']
cols_to_int_zero = ['Quantidade', 'Data da transação']

for col in cols_to_fill_nao_informado:
    df[col] = df[col].fillna('Nao Informado')

for col in cols_to_numeric_median:
    df[col] = df[col].astype(str).str.replace(',', '.')
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(df[col].median())

for col in cols_to_int_zero:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)

print(df.info())
```
