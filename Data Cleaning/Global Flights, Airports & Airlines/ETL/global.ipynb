```python
import pandas as pd

df_airlines = pd.read_csv('/home/stefanyfranca/Área de trabalho/Discovering-pandas/Data Cleaning/Global Flights, Airports & Airlines/Data/airlines.csv')
df_airports = pd.read_csv('/home/stefanyfranca/Área de trabalho/Discovering-pandas/Data Cleaning/Global Flights, Airports & Airlines/Data/airports.csv')
df_routes = pd.read_csv('/home/stefanyfranca/Área de trabalho/Discovering-pandas/Data Cleaning/Global Flights, Airports & Airlines/Data/routes.csv')
```

```python
#Filtrar aeroportos dos EUA (country)
american_airports = df_airports[df_airports['iso_country'] == 'US']
print(american_airports)

print("=================================================================")

#Contar quantos existem por estado 
american_airports_states = american_airports.groupby('iso_region').size().reset_index(name='count')
print(american_airports_states)
```

```python
#verificar quantas linhas estão vazias por coluna
vazios_routes = df_routes.isnull().sum()
print(vazios_routes)
```

```python
#subistituir valores vazios por 'Unknown'
df_routes_preenchido = df_routes.fillna("Unknown")
#conferir
vazios_routes = df_routes_preenchido.isnull().sum()
print(vazios_routes)
```

```python
#filtramos apenas ativas E que tenham ID maior que 1 
filtro_reais = (df_airlines['active'] == 'Y') & (df_airlines['airline_id'] > 1)

antigas_ativas = df_airlines[filtro_reais].sort_values('airline_id').head(10)
print(antigas_ativas)
```

```python
#encontrar os 5 aeroportos de origem (source airport) que possuem o maior volume de rotas.
top5 = df_routes['src_airport'].value_counts().head(5).reset_index()
print(top5)
```

```python
#Identificar qual modelo de aeronave (equipment) é o mais utilizado em todas as rotas.
equip_individual = df_routes['equipment'].dropna().str.split().explode()

top_individual = equip_individual.value_counts().head(10).reset_index()

print(top_individual)
```

```python
#Criar um ranking de países pelo número de companhias aéreas registradas neles.
topPaises = df_airlines['country'].value_counts().head(10)
print(topPaises)
```

```python
df_merged = pd.merge(
    df_routes, 
    df_airlines[['iata_code', 'active', 'callsign']], 
    left_on='airline_iata',                           
    right_on='iata_code',                             
    how='left'                                        
)

print(df_merged[['airline_name', 'airline_iata', 'active']].head())
```

```python
# criar coluna is_international. Comparar se o país do aeroporto de origem é o mesmo do destino.
df_merged['is_international'] = df_merged['src_country'] != df_merged['dst_country']

contagem_voos = df_merged['is_international'].value_counts()

print("Distribuição de voos:")
print(contagem_voos)

print("\nPorcentagem:")
print(df_merged['is_international'].value_counts(normalize=True) * 100)
```
