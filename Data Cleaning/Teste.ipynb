```python
import pandas as pd
import numpy as np

Dados = {
                'Produtos'  : ['Sabonete','Shampoo','Condicionador','Esponja'],
                'Categoria' : ['Higiene Corporal', 'Capilar', 'Capilar', 'Cozinha'],
                'Preço'     : [5,22,24,2.5],
                'Quantidade' : [20,7,9,15]
            }

df_vendas = pd.DataFrame(Dados)
print(Dados)
```

```python
produtos_capilares = df_vendas[df_vendas['Categoria'] == 'Capilar']
print(produtos_capilares)

produtos_baratos = df_vendas[df_vendas['Preço'] < 10]
print("\n", produtos_baratos)

estoque_cheio = df_vendas[df_vendas['Quantidade'] > 15]
print("\n", estoque_cheio)

produto_especifico = df_vendas[df_vendas['Produtos'] == 'Shampoo']
print("\n", produto_especifico)
```

```python
#dataframe apenas com as colunas de produto e preço
df_reduzido = df_vendas[['Produtos', 'Preço']]
print(df_reduzido)
```

```python
#criar linha de faturamento
df_vendas['Faturamento Total'] = df_vendas['Preço']*df_vendas['Quantidade']
print(df_vendas)
#criar linha de preço com 10% de desconto
df_vendas['Preço com Desconto'] = df_vendas['Preço']*(1 - (10/100))
print('\n',df_vendas)
```

```python
#faturamento por categoria
faturamento_categoria = df_vendas.groupby('Categoria')['Faturamento Total'].sum().reset_index()
print(faturamento_categoria)
```

```python
#media de preços dos produtos
media = df_vendas['Preço'].mean()
print(media)
#media por categoria
media_por_categoria = df_vendas.groupby('Categoria')['Preço'].mean().reset_index()
print(media_por_categoria)
```

```python
#faturamento em ordem decrescente
top_faturamento = df_vendas.sort_values('Faturamento Total', ascending= False)
print(top_faturamento)
```

```python
#top 2 produtos com maior faturamento
top_produtos = top_faturamento.head(2)
print(top_produtos)
```

```python
#Todos os produtos em letra maiúscula
df_vendas['Produtos'] = df_vendas['Produtos'].str.upper()
print(df_vendas)
```

```python
#verificar se tem dados nulos no dataframe
numeros_nulos = df_vendas.isnull().sum()
print(numeros_nulos)
```

```python
#Criar a coluna Status
df_vendas['Status'] = ''
print(df_vendas)
```

```python
#Adicionar informação a coluna Status
df_vendas['Status'] = np.where(df_vendas['Quantidade'] > 10, 'Estoque Alto', 'Estoque Baixo')
print(df_vendas)
```

```python
df_vendas['Preço Promocional'] = np.where(df_vendas['Status'] == 'Estoque Alto', df_vendas['Preço com Desconto']*(1-(5/100)), df_vendas['Preço com Desconto'])
print(df_vendas)
```
